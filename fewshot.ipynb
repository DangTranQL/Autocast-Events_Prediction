```python
def is_float(string):
    try:
        float(string)
        return True
    except ValueError:
        return False
def is_int(string):
    try:
        int(string)
        return True
    except ValueError:
        return False
```

```python
import string

# Creating a hash table
# getting the overall index of input_ids that contains the correct answer

# First, let's create a dictionary to associate A,B,C,... to 0,1,2,...

MC_Dict = {}

MC_lower = {}

for index, alphabet in enumerate(string.ascii_uppercase):
    MC_Dict[alphabet] = index
    
for index, alphabet in enumerate(string.ascii_lowercase):
    MC_lower[alphabet] = index
```
